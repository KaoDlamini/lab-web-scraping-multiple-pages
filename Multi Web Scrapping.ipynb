```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import math
```

```python
r = requests.get('https://www.acclaimedmusic.net/year/1990-99s.htm')
r.status_code
```

```python
r =r.content
soup = BeautifulSoup(r, 'html.parser')
print(soup.prettify())
```

```python
table=soup.find_all('table', attrs={'class': "center"})[0]
```

```python
artists=[]
for row in table.find_all('tr')[1:]:
    artist= row.find_all('td')
    
    if len(artist)>=2:
        artist= artist[2].get_text(strip=True) 
        artists.append(artist)

artists
```

```python
len(artists)
```

```python
songs=[]
for row in table.find_all('tr')[1:]:
    song= row.find_all('td')
    
    if len(song)>=2:
        song= song[3].get_text(strip=True) 
        songs.append(song)

songs
```

```python
len(songs)
```

```python
all_artists = []
all_songs = []

decade_ranges = ["1990-99", "2000-09", "2010-19"]

for i in decade_ranges:
    r = requests.get(f"https://www.acclaimedmusic.net/year/{i}s.htm")
    soup = BeautifulSoup(r.content, 'html.parser')
    table=soup.find_all('table', attrs={'class': "center"})[0]
    
    for row in table.find_all('tr')[1:]:
        artist_data= row.find_all('td')
    
        if len(artist_data)>=3:
            artist= artist_data[2].get_text(strip=True) 
            all_artists.append(artist)
            songs= artist_data[3].get_text(strip=True) 
            all_songs.append(songs)
```

```python
len(all_artists)
```

```python
len(all_songs)
```
